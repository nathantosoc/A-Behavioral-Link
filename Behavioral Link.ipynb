## Cleaning Youth Smoking and Drug Dataset

```python
## Importing Dataset
file_path = 'youth_smoking_drug_data_10000_rows_expanded.csv'
data = pd.read_csv(file_path)

#Drop duplicate rows
data.drop_duplicates(inplace=True)

#Drop missing values
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)

#Select relevant columns
data = data[['Age_Group','Gender','Smoking_Prevalence','Drug_Experimentation','Socioeconomic_Status','Mental_Health']]

#Standardize Text
data['Gender'] = data['Gender'].str.strip().str.capitalize()
data['Age_Group'] = data['Age_Group'].str.strip()
data['Socioeconomic_Status'] = data['Socioeconomic_Status'].str.strip().str.title()

#Standardize the Age Range into an average
data["Age_Group"].value_counts()

def standardize_age(string):
    age = string.lower().strip()
    if "10-14" == age:
        output = 12
    elif "20-24" == age:
        output = 22
    elif "60-69" == age:
        output = 64
    elif "15-19" == age:
        output = 17
    elif "40-49" == age:
        output = 44
    elif "50-59" == age:
        output = 54
    elif "80+" == age:
        output = 80
    elif "25-29" == age:
        output = 27
    elif "70-79" == age:
        output = 74
    else:
        output = np.nan
    return output

data["Standardized_Age"] = data["Age_Group"].apply(standardize_age)

#View cleaned data
data.head()
```

I loaded the dataset by specifying the file path and using the read_csv() function from the pandas library. This reads the CSV file and stores the data in a DataFrame named df, which allows for easy manipulation and analysis. Then started by removing any duplicate rows from the dataset using the drop_duplicates() function. This ensures that each record in the DataFrame is unique, preventing repeated entries from skewing the analysis. I set inplace=True to apply the change directly to the DataFrame without needing to reassign it. Using df.dropna(inplace=True), I remove any rows that contain missing data to maintain the quality and integrity of the dataset. After dropping these rows, I reset the index with df.reset_index(drop=True, inplace=True) to keep the DataFrame well-organized and avoid gaps in the index sequence. I standardize columns that contain Yes/No responses to ensure consistency in analysis. In this case, I focus on the 'School_Programs' column. Using the applymap() function, I convert all 'Yes' values to 1 and all other responses to 0. I standardize the formatting of text data to ensure consistency across the dataset. For the 'Gender' column, I remove any leading or trailing spaces and capitalize the first letter of each value. In the 'Age_Group' column, I strip any extra whitespace and standerdized the age. For 'Socioeconomic_Status', I strip whitespace and convert the text to title case.

## Section 1 - Analysis of Drug Use & Youth Smoking
This section of my EDA explores the relationships and correlation between substance use and socioeconomic status. This dataset focuses particularly on smoking prevalence and drug experimentation and its relationship with the users' socioeconomic status (categorized into high, middle, and low). Smoking prevalence is measured through smoking status (likeliehood of smoking) and drug experimentation is a measure of whether the individual has experiented with drugs (general).

```python
#Regression Test on Drug Experimentation and Age
mod_1 = smf.ols(formula = 'Drug_Experimentation ~ Standardized_Age', data = data)
res_1 = mod_1.fit()
res_1.summary()
```

The p-value for this regression is statistically significant under the standard critical value, implying a possible negative correlation between drug-usage and age, which could hints towards the possibly that as people mature, they typically consume less drugs. The analysis also informs me that if my hypothesis is true, people should experience positive social and economic benefits as their life goes on.

```python
mod_1 = smf.ols(formula = 'Mental_Health ~ Drug_Experimentation', data = data)
res_1 = mod_1.fit()
res_1.summary()
```

Overall, drug experimentation leads to better mental health over time (slightly) with the coefficient of .0043. This coefficient doesn't tell me much about the direct relationship between the variables, however it is statistically significant. This implies there isn't any evidence in the contrary to my hypothesis, and that I should continue the relationships between variables in my models.

```python
print(data["Drug_Experimentation"].min())
print(data["Drug_Experimentation"].max())
print(data["Drug_Experimentation"].mean())
print(data["Drug_Experimentation"].std())
```

Now I will divide the drug_experimentation levels with buckets going from min to max, with the bucket size at half the standard deviation. I am doing this to see the frequency of different social classes within each drug usage group, and see if I can extrapolate a correlation there.

```python
def drug_levels(drug_experimentation):
    drug_l = float(drug_experimentation)
    if drug_l > 10 and drug_l < 18.75:
        return "Group 0 Users"
    elif drug_l > 18.75 and drug_l < 27.5:
        return "Group 1 Users"
    elif drug_l > 27.5 and drug_l < 36.25:
        return "Group 2 Users"
    elif drug_l > 36.25 and drug_l < 45:
        return "Group 3 Users"
    elif drug_l > 45 and drug_l < 53.75:
        return "Group 4 Users"
    elif drug_l > 53.75 and drug_l < 62.5:
        return "Group 5 Users"
    elif drug_l > 62.5 and drug_l < 70:
        return "Group 6 Users"
    else:
        return np.nan
    
data["Drug_Groups"] = data["Drug_Experimentation"].apply(drug_levels)

data.head()
data["Drug_Groups"].value_counts()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data = data, x = "Drug_Groups", hue = "Socioeconomic_Status")
```

In the plot it can be seen that Group 0 users have the most amount of high class users, however a direct comparison may not provide a tangible statistical inference. As seen in the previous value counts the amount of users taper off. However, I can compare my classes within themselves. This can be done by standardizing the classes by proportion, which will give a more accurate idea of what the data distribuition.

Head counts decline as it progresses from group 0 to group 4 simply as fewer respondents in this data set fell within the higher-prevalence bins. Looking at the individual bins of users, the blue, orange, and green bars representing the difference socioeconomic statuses of the respondents varies minimally suggesting there may not be a clear socioeconomic skew.

```python
#Creating a proportionalized data set with the proportion column for each socieconomic status
counts = data.groupby(['Drug_Groups', 'Socioeconomic_Status']).size().reset_index(name = 'Count')
counts['proportion'] = counts.groupby('Drug_Groups')['Count'].transform(lambda x: x / x.sum())
plt.figure(figsize=(14, 8))
plt.xticks(rotation=45)
plt.title("Proportions of Socioeconomic Status within Each Drug Group")
sns.barplot(data=counts, x = 'Drug_Groups', y = 'proportion', hue = "Socioeconomic_Status")
```

After having normalized each column, the assumptions that could be made visually from the previous plot seem to be confirmed in this graph. There is no significant skew of drug users across the socioeconomic tiers across all the different bins of users. The minor differences (~2%) can possibly be accounted for within sampling noise for this 10,000 sample dataset. 

```python
mod_2 = smf.ols(formula = 'Smoking_Prevalence ~ Socioeconomic_Status', data = data)
res_2 = mod_2.fit()
res_2.summary()
```

Smoking prevalence averages around 27% across all socioeconomic levels in this dataset. Neither "low" nor "middle" socioeconomic levels differ significantly from the high socioeconomic sample (0.08% and 0.37% respectively). with a minimal R-squared value, socioeconomic status may not be a useful predictor of smoking prevalence among adults. 

```python
print(data["Smoking_Prevalence"].min())
print(data["Smoking_Prevalence"].max())
print(data["Smoking_Prevalence"].mean())
print(data["Smoking_Prevalence"].std())
```

```python
def smoking_levels(smoking_rate):
    rate = float(smoking_rate)
    if 10 < rate <= 18.75:
        return "Group 0 Smokers"
    elif 18.75 < rate <= 27.5:
        return "Group 1 Smokers"
    elif 27.5 < rate <= 36.25:
        return "Group 2 Smokers"
    elif 36.25 < rate <= 45:
        return "Group 3 Smokers"
    elif 45 < rate <= 53.75:
        return "Group 4 Smokers"
    elif 53.75 < rate <= 62.5:
        return "Group 5 Smokers"
    elif 62.5 < rate <= 70:
        return "Group 6 Smokers"
    else:
        return np.nan

data["Smoking_Groups"] = data["Smoking_Prevalence"].apply(smoking_levels)

data["Smoking_Groups"].value_counts()
```

```python
plt.figure(figsize = (14, 8))
sns.countplot(data = data, x = "Smoking_Groups", hue = "Socioeconomic_Status")
plt.title("Socioeconomic Status within Smoking Groups")
plt.xlabel("Smoking Group")
plt.ylabel("# of Individuals")
plt.xticks(rotation = 45)
plt.legend(title = "Socioeconomic Status")
plt.show()
```

Within each of these smoking groups, it does not appear that smoking prevalence leans significantly towards a specific socioeconomic status. To verify this, the data will be standardized to proportions to provide a more accurate representation of each smoking group. 

```python
smoke_counts = (
    data
    .groupby(["Smoking_Groups", "Socioeconomic_Status"])
    .size()
    .reset_index(name = "Count")
)
smoke_counts["proportion"] = (
    smoke_counts
    .groupby("Smoking_Groups")["Count"]
    .transform(lambda x: x / x.sum())
)

plt.figure(figsize=(14, 8))
sns.barplot(
    data = smoke_counts,
    x = "Smoking_Groups",
    y = "proportion",
    hue = "Socioeconomic_Status"
)
plt.title("Proportions of Socioeconomic Status within Smoking Groups")
plt.xlabel("Smoking Group")
plt.ylabel("Proportion within Group")
plt.xticks(rotation = 45)
plt.legend(title = "Socioeconomic Status")
plt.show()
```

Once again, across the difference smoking groups, it appears that there a minor differences of smoking prevalence across the three socioeconomic statuses but remains difficult to establish a strong correlation between socioecnomic status and smoking prevalence.

Across the data sample of 10,000 respondents, it appears that smoking intensity and general drug usage/experimentation is largely independent of the user's socioeconomic status.

## Section 2 - Data Analysis & Results Regarding Personality and Drug Use

This analysis explores how personality traits (neuroticism, extraversion, openness) relate to the usage patterns of common drugs (cannabis, alcohol, nicotine) among individuals. I also explore potential confounding factors such as age, gender, and country.


### 1. Distribution of Drug Use EDA

```python
for drug in ['Cannabis', 'Alcohol', 'Nicotine']:
    plt.figure(figsize=(6, 4))
    sns.countplot(x=drug, data=df)
    plt.title(f'{drug} Use Frequency')
    plt.show()
```

This chart shows the distribution of self-reported cannabis, alcohol, and nicotine use frequencies across the sample. **Cannabis** and **nicotine** both display a **bimodal distribution**, with many participants either abstaining completely (frequency = 0) or using heavily (frequency = 6), suggesting **polarized usage habits**. **Alcohol use**, by contrast, is more **skewed toward frequent use**, with a large number of participants reporting consumption levels of 5 or 6. This indicates that **alcohol is more commonly and consistently used** than the other two substances. These patterns highlight key differences in **substance norms and usage behaviors** that may influence or be influenced by **personality and socioeconomic traits**.

### 2. Personality Trait Distributions

```python
for trait in ['Neuroticism', 'Extraversion', 'Openness']:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[trait], bins=20, kde=True)
    plt.title(f'Distribution of {trait}')
    plt.show()
```

This figure displays the distributions of **Neuroticism**, **Extraversion**, and **Openness** scores reported by participants. The distributions are approximately normal, though **Neuroticism** and **Openness** appear to have slightly wider spreads, suggesting more variability in how individuals experience emotional instability and openness to new ideas. **Extraversion**, by contrast, is more narrowly centered, indicating that social tendencies may be more consistently reported. These distribution shapes imply that while most individuals fall near the mid-range for all three traits, there are meaningful subsets with high or low trait levels. Understanding these personality score distributions is essential before analyzing their potential relationships with drug use, as skew or variance in personality traits could influence how they correlate with behavioral outcomes.

### 3. Personality vs Drug Use

```python
for drug in ['Cannabis', 'Alcohol', 'Nicotine']:
    # Reshape the data so trait columns become a single column
    melted = df.melt(id_vars=[drug], value_vars=['Neuroticism', 'Extraversion', 'Openness'],
                     var_name='Trait', value_name='Score')
    
    plt.figure(figsize=(8, 5))
    sns.lineplot(x=drug, y='Score', hue='Trait', data=melted, estimator='mean', errorbar=None, marker='o')
    
    plt.title(f'Mean Trait Scores by {drug} Use')
    plt.xlabel(f'{drug} Use')
    plt.ylabel('Mean Trait Score')
    plt.legend(title='Trait')
    plt.tight_layout()
    plt.show()
```

These line plots display the average scores for three personality traits—Neuroticism, Extraversion, and Openness across varying levels of self-reported use of Cannabis, Alcohol, and Nicotine. The x-axis reflects frequency of use for each drug, while the y-axis represents the mean personality trait score. A general trend observed for Openness is that individuals who report more frequent cannabis and nicotine use tend to score higher, suggesting that more open individuals may be more likely to experiment with or routinely use these substances. In contrast, Neuroticism shows a less consistent relationship, with a modest upward trend in nicotine users, possibly indicating emotional vulnerability in heavier users. Extraversion appears mostly stable across all three drug types, implying weaker associations. These patterns highlight how certain personality dimensions may be more predictive or reactive to substance use behaviors, and underscore the importance of examining multiple psychological traits in behavioral health studies.

```python
education_values = ['No HS', 'HS', 'College/Cert', 'Some College', 'Bachelor', 'Graduate']
for drug in ['Cannabis', 'Alcohol', 'Nicotine']:
    plt.figure(figsize=(8, 5))

    # Calculate mean usage by education level
    grouped = df.groupby('Education', observed = False)[drug].mean().reindex(education_values)

    sns.lineplot(
        x=grouped.index,
        y=grouped.values,
        marker='o'
    )

    plt.title(f'Mean {drug} Use by Education Level')
    plt.xlabel('Education Level')
    plt.ylabel(f'Mean {drug} Use ')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
